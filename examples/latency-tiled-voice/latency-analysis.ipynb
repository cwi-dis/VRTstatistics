# Latency analysis notebook

Plots per-component latency contributions, end-to-end latency, frame rates, and point counts
from a two-machine latency experiment (PointCloud sender + Avatar receiver).
Use this after running `VRTstatistics-ingest -a latency` with the `latency` example config.

Run Jupyter from your experiment directory (the one containing `run-*/` subdirectories):
```
cd /path/to/my-experiment
source /path/to/VRTstatistics/.venv/bin/activate
jupyter lab
```

````
import glob
import os

# Set to a specific run directory, or leave as None to use the most recent one.
run_dir = None

if run_dir is None:
    candidates = sorted(glob.glob("run-*"))
    if not candidates:
        raise FileNotFoundError("No run-* directories found. Run VRTstatistics-ingest first.")
    run_dir = candidates[-1]

combined_json = os.path.join(run_dir, "combined.json")
print(f"Using: {combined_json}")
````

````
%matplotlib inline
from VRTstatistics.datastore import DataStore
from VRTstatistics.plots import plot_latencies, plot_framerates_and_dropped, plot_pointcounts

ds = DataStore(combined_json)
ds.load()
````

````
plot_latencies(ds);
````

````
plot_framerates_and_dropped(ds);
````

````
plot_pointcounts(ds);
````
