# Simple sanity-check notebook

Plots CPU, memory, and bandwidth from a two-machine SimpleAvatar session.
Use this after running `VRTstatistics-ingest` with the `simple` example config
to verify the end-to-end pipeline is working.

Run Jupyter from your experiment directory (the one containing `run-*/` subdirectories):
```
cd /path/to/my-experiment
source /path/to/VRTstatistics/.venv/bin/activate
jupyter lab
```

````python
import glob
import os

# Set to a specific run directory, or leave as None to use the most recent one.
run_dir = None

if run_dir is None:
    candidates = sorted(glob.glob("run-*"))
    if not candidates:
        raise FileNotFoundError("No run-* directories found. Run VRTstatistics-ingest first.")
    run_dir = candidates[-1]

combined_json = os.path.join(run_dir, "combined.json")
print(f"Using: {combined_json}")
````

````python
%matplotlib inline
from VRTstatistics.datastore import DataStore
from VRTstatistics.plots import plot_resources

ds = DataStore(combined_json)
ds.load()
plot_resources(ds);
````
